```python
import pandas as pd
import numpy as np
import os

np.random.seed(42)

n = 100

data = {
    'id': np.arange(1, n+1),
    'neighbourhood': np.random.choice(['Downtown', 'Uptown', 'Suburbs'], n),
    'room_type': np.random.choice(['Entire home/apt', 'Private room', 'Shared room'], n, p=[0.6, 0.3, 0.1]),
    'price': np.random.randint(25, 300, n),
    'minimum_nights': np.random.randint(1, 8, n),
    'number_of_reviews': np.random.randint(0, 200, n),
    'last_review': pd.to_datetime(np.random.choice(pd.date_range('2023-01-01', '2023-10-31', freq='D'), n)),
    'reviews_per_month': np.round(np.random.uniform(0.0, 10.0, n), 2),
    'calculated_host_listings_count': np.random.randint(1, 5, n),
    'availability_365': np.random.randint(0, 366, n),
    'review_scores_rating': np.random.randint(60, 100, n)
}

df = pd.DataFrame(data)
df.to_csv('data/listings_dummy.csv', index=False)
df.head()
```

```python
# Load the saved file — this is how you will always start
df = pd.read_csv('data/listings_dummy.csv')

# Shape — how many rows and columns exist
print("Shape:", df.shape)

# Column names and their data types
print("\nData types:")
print(df.dtypes)

# Basic info — combines both above plus memory usage
print("\nFull info:")
df.info()
```

```python
# Convert neighbourhood and room_type to category
df['neighbourhood'] = df['neighbourhood'].astype('category')
df['room_type'] = df['room_type'].astype('category')

# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'])

# Verify the fix
print(df.dtypes)
```
